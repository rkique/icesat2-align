```python
import h5py
import pandas as pd
import numpy as np


def getATL03(f, beam):
    attributes = {'bckgrd_rate': '/bckgrd_atlas/bckgrd_rate'}
    df03 = pd.DataFrame({})
    for name, attr in attributes.items():
        df03[f'{name}'] = f[beam+f'{attr}'][:]
    return df03

def getATL07(f,beam):
    attributes = {'lats':'/sea_ice_segments/latitude', 
                  'lons':'/sea_ice_segments/longitude', 
                  'heights':'/sea_ice_segments/heights/height_segment_height',
                  'dt':'/sea_ice_segments/delta_time',
                  'conf': '/sea_ice_segments/heights/height_segment_confidence',
                  'stype': '/sea_ice_segments/heights/height_segment_type',
                  'ssh_flag': '/sea_ice_segments/heights/height_segment_ssh_flag',
                  'gauss': '/sea_ice_segments/heights/height_segment_w_gaussian',
                  'photon_rate': '/sea_ice_segments/stats/photon_rate',
                  'cloud': '/sea_ice_segments/stats/cloud_flag_asr',
                  'backgr_r_200': '/sea_ice_segments/stats/backgr_r_200',
                  'backgr_r_25': '/sea_ice_segments/stats/backgr_r_25',
                  'background_r_norm': '/sea_ice_segments/stats/background_r_norm',
                  'mss': '/sea_ice_segments/geophysical/height_segment_mss',
                  'ocean': '/sea_ice_segments/geophysical/height_segment_lpe',
                  'lpe': '/sea_ice_segments/geophysical/height_segment_lpe',
                  'ib': '/sea_ice_segments/geophysical/height_segment_ib',
                  'width_of_fit': '/sea_ice_segments/heights/height_segment_w_gaussian',
                  'solar_elevation': '/sea_ice_segments/geolocation/solar_elevation',
                  }
    df07 = pd.DataFrame({})
    for name, attr in attributes.items():
        df07[f'{name}'] = f[beam+f'{attr}'][:]

    #information not specific to the gt3r beam
    print(f['ancillary_data/surface_classification/theta_cntl'][:][0])
    print(f['ancillary_data/surface_classification/theta_nlb'][:][0])
    print(f['ancillary_data/surface_classification/theta_ref'][:][0])
    print(f['ancillary_data/surface_classification/beam_gain'][:])
    
    vals = ['p1', 'p2', 'p3', 'p4', 'w1', 'w2', 'b1']
    for val in vals:
        path = f'/ancillary_data/surface_classification/{val}'
        print(f'{val}: {f[path][:][0]}')
    return df07
```

```python
f = h5py.File('../base_data/ATL07-01_20220726161113_05311601_006_02.h5', 'r')
beam='gt3r'

#f.keys() will show the dictionary
df07=getATL07(f, beam)
df07
```

```python
df07.to_csv('gt3r_full.csv', index=False)
```

```python
import h5py
import pandas as pd
import numpy as np

f = h5py.File('../base_data/20220726_RGT531_QD064_FL1_SC_TD_UTM16_revised_geocorr-applied_v2.h5', 'r')

beam='gt3r'
#beam='gt3r'

output_csv_file = '../base_data/Chirop_20220726_FL1_SC_GT3R.csv'
# geocorrected chiroptera format has the following keys: ['GPS_time', 'chiroptera_z', 'corr_height_rel005', 'corr_height_rel006', 'geophysical', 'latitude', 'longitude']

def getCorrectedChiroptera(f):
    GPS_time = f['GPS_time']
    chiroptera_z = f['chiroptera_z']
    corr_height_rel005 = f['corr_height_rel005']
    corr_height_rel006 = f['corr_height_rel006']
    geophysical_ib_rel006 = f['geophysical']
    latitude = f['latitude']
    longitude = f['longitude']

    print(f"Length of GPS_time: {len(GPS_time)}")
    print(f"Length of chiroptera_z: {len(chiroptera_z)}")
    print(f"Length of corr_height_rel005: {len(corr_height_rel005)}")
    print(f"Length of corr_height_rel006: {len(corr_height_rel006)}")
    print(f"Length of latitude: {len(latitude)}")
    print(f"Length of longitude: {len(longitude)}")

    chiroptera = pd.DataFrame({
        'GPS_time': GPS_time,
        'chiroptera_z': chiroptera_z,
        'corr_height_rel005': corr_height_rel005,
        'corr_height_rel006': corr_height_rel006,
        'latitude': latitude,
        'longitude': longitude,
    })
    
    return chiroptera

#df07= rd.getATL07(f,beam)
correctedChiroptera = getCorrectedChiroptera(f)
correctedChiroptera.to_csv(output_csv_file, index=False)
```
