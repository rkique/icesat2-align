```python
import csv
file = open('data/gt3r_full.csv')
csvreader = csv.reader(file)
result = []
count = -1
current_water_ind = -5
for row in csvreader:
    if row[5] == '1':
        current_water_ind = count
    if row[5] != '1':
        if current_water_ind == count - 1:
            result.append(current_water_ind)
    count += 1
        
print(result)
```
